```python
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate

plt.style.use('ggplot')


x_axis = np.linspace(0, 10, 8)
y_axis = np.sin(x_axis)

func = interpolate.interp1d(x_axis, y_axis, kind='cubic')

x_interplt = np.linspace(0, 10, 1000)
y_interplt = func(x_interplt)

plt.figure()
plt.plot(x_axis, y_axis, 'o')
plt.plot(x_interplt, y_interplt)


```
