# Read data

# Plotting graph with seaborn

```python
plt.figure(figsize=(6,8))
seaborn.set(font_scale = 2.5)

#Counting the attribute class from raw_diabetes_data
seaborn.countplot(x = 'class', data=raw_diabetes_data, palette="Set1")

plt.ylabel('People')
axis_labels = ['Diabetic', 'Non Diabetic']
x_axis = [0, 1] 
plt.xticks(x_axis, axis_labels, rotation ='horizontal') 
```
